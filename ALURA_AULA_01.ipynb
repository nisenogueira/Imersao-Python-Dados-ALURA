#Aula 01 - Explore Dados com Pandas

```python
# importando a biblioteca Pandas para o projeto

import pandas as pd
```

```python
# O primeiro passo é importar a base de dados que vamos utilizar no projeto. O link utilizado está disponivel no repositorio da Alura.

df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
```

```python
# Iniciamos visualizado as primeiras linhas da base de dados.

df.head()
```

```python
# O método .info() fornece um resumo conciso do DataFrame, incluindo o número de entradas, o número de colunas, os tipos de dados de cada coluna e a quantidade
# de valores não nulos.

df.info()
```

```python
# O método .describe() é utilizado para gerar estatísticas descritivas de uma base de dados. Ele fornece um resumo estatístico das colunas numéricas, incluindo contagem,
# média, desvio padrão, valores mínimo e máximo, e os percentis.

df.describe()
```

```python
# Já o atributo .shape em Python é utilizado para obter as dimensões de uma base de dados. Ele retorna uma tupla que contém o número de linhas e o número de colunas.

df.shape
```

```python
# Criando a variavel 'linha' e 'coluna' pegando o primeiro e o segundo item de 'shape'.

linhas, colunas = df.shape[0], df.shape[1]
print("linhas:", linhas)
print("colunas:", colunas)
```

```python
# Visualizando o nome de todas as colunas presentes na base de dados.

df.columns
```

```python
# Agora vamos comecar a melhorar a visualização dos dados, renomeando as colunas para nomes mais descritivos em Português.

df.rename(columns={
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_in_usd': 'moeda',
    'salario_em_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}, inplace=True)
df.columns
```

```python
# Vamos verificar se as alterações foram feitas corretamente.

df.head()
```

```python
# Começamos agora a fazer algumas analises.
# O método df["senioridade"].value_counts() é utilizado para contar a frequência de cada valor único na coluna "senioridade" da base de dados.
# Isso é útil para entender a distribuição dos níveis de experiência dos funcionários.
# Vale lembrar que esse metodo calcula apenas dados numéricos.

df["senioridade"].value_counts()
```

```python
# Faremos o mesmo com algumas outras colunas que são relevantes para essa analise.

df["contrato"].value_counts()
```

```python
# Aqui contaremos quantas pessoas trabalham em cada categoria da coluna 'remoto' [0 = trabalho presencial, 100 = remoto, 50 = hibrido]

df["remoto"].value_counts()
```

```python
# O total de pessoas que trabalham em cada tipo de contrato de trabalho ['full time, contrato, part time, freelancer']

df["contrato"].value_counts()
```

```python
# A contagem do tamanho das empresas que estão na base de dados. [Medium, Large, Small]

df["tamanho_empresa"].value_counts()
```

```python
# O total de cada categoria de senioridade

df['senioridade'] = df['senioridade'].replace({
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
})
df['senioridade'].value_counts()
```

```python
# Podemos também traduzir os dados dentro das colunas para facilitar a leitura.

df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
})
df['contrato'].value_counts()
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
})
df['tamanho_empresa'].value_counts()
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
})
df['remoto'].value_counts()
```

```python
# Mais uma vez vamos verificar se as modificações foram feitas

df.head()
```

```python
# Como ultima analise desta primeira aula, vamos agora utilizar o metodo .describe(include='object')
# que é utilizado para gerar estatisticas descritivas de colunas não numéricas (geralmente strings)

df.describe(include='object')
```
