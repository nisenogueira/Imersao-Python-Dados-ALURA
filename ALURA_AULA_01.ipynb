```python
df.columns
```

```python
df.rename(columns={
    'ano': 'ano',
    'nivel_experiencia': 'senioridade',
    'tipo_emprego': 'contrato',
    'cargo': 'cargo',
    'salario': 'salario',
    'moeda_salario': 'moeda',
    'salario_em_usd': 'usd',
    'residencia_empregado': 'residencia',
    'taxa_remoto': 'remoto',
    'local_empresa': 'empresa',
    'tamanho_empresa': 'tamanho_empresa'
}, inplace=True)
df.columns
```

```python
df.head()
```

```python
df["senioridade"].value_counts() # conta os valores de cada categoria em 'senioridade'
```

```python
df["contrato"].value_counts() # conta os valores de cada categoria em 'contrato'
```

```python
df["remoto"].value_counts() # conta os valores de cada categoria em 'remoto' [0 = trabalho presencial, 100 = remoto, 50 = hibrido]
```

```python
df["contrato"].value_counts() # conta os valores de cada categoria em 'contrato'
```

```python
df["tamanho_empresa"].value_counts() # conta os valores de cada categoria em 'tamanho_empresa'
```

```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
})
df['senioridade'].value_counts()
```

```python
df['contrato'] = df['contrato'].replace({
    'Tempo Integral': 'Tempo Integral',
    'Contrato': 'Contrato',
    'Meio Perodo': 'Meio Perodo',
    'Freelancer': 'Freelancer'
})
df['contrato'].value_counts()
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Mdio',
    'L': 'Grande',
    'S': 'Pequeno'
})
df['tamanho_empresa'].value_counts()
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    100: 'Remoto',
    50: 'Hbrido'
})
df['remoto'].value_counts()
```
